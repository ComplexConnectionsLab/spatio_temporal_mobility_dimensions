```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from scipy.stats import zscore
from aux_functions import *

%load_ext autoreload
%autoreload 2
```

```python
# Loads the table previous generated and saved
devices_per_day = pd.read_csv('files/devices_per_day_complete_with_soc.csv')
devices_per_day.head()
```

```python
colors_y = ['#595959', '#ff5d8f']

sns.set(font_scale=2)
sns.set_style('white')
data = devices_per_day[devices_per_day.week<53].groupby(['year', 'week', 'geo_code']).mean().reset_index()
f, ax = plt.subplots(1, figsize=(16, 5)) 
sns.lineplot(data=data, x="week", hue='year', y="prop_norm", linewidth=3, palette=colors_y)
plt.axvline(x=13,linestyle='dashed',lw=3,color='#595959',label='1st Lockdown')
plt.xlabel("Week of the year")
plt.ylabel("Number of Trips")
sns.move_legend(ax, "upper center", ncol=2, title=None, frameon=False,)
plt.show()
```

```python
cols_ = ['week', 'prop_norm']
diff_df = devices_per_day[devices_per_day.year==2019].groupby(cols_[:1]).mean().reset_index()[cols_]
diff_df.rename(columns={"prop_norm": "2019"}, inplace=True)

temp = devices_per_day[(devices_per_day.year==2020)].groupby(cols_[:1]).mean().reset_index()[cols_]
temp.rename(columns={"prop_norm": "2020"}, inplace=True)

diff_df = diff_df.merge(temp, on=['week'])
diff_df['change'] = (diff_df["2020"].values - diff_df["2019"].values)/diff_df["2020"].values
diff_df = diff_df.round(4)
diff_df.head()
```

```python
sns.set(font_scale=2)
sns.set_style('white')
f, ax = plt.subplots(1, figsize=(16, 5)) 
sns.lineplot(data=diff_df, x="week", y='change', linewidth=3, color=colors_y[0])
plt.axvline(x=13,linestyle='dashed',lw=3,color='#595959',label='1st Lockdown')
plt.xlabel("Week of the year")
plt.ylabel("Baseline (%)")
plt.savefig("figures/change_in_visits_time_series.svg", transparent=True)
plt.show()
```

```python
interval = 4
file_name = "figures/duration_trips.svg"
plot_lollipop(diff_df, interval, file_name)
```

```python
cols_ = ['week', 'prop_norm']
diff_df_ru = devices_per_day[devices_per_day.year==2019].groupby(cols_[:1]).mean().reset_index()[cols_]
diff_df_ru.rename(columns={"prop_norm": "2019"}, inplace=True)

temp = devices_per_day[(devices_per_day.year==2020)].groupby(cols_[:1]).mean().reset_index()[cols_]
temp.rename(columns={"prop_norm": "2020"}, inplace=True)
diff_df_ru = diff_df_ru.merge(temp, on=['week'])
diff_df_ru['change'] = (diff_df_ru["2020"].values - diff_df_ru["2019"].values)/diff_df_ru["2020"].values

temp_rural = devices_per_day[(devices_per_day.year==2020) & (devices_per_day.NRUC11=='Rural')].groupby(cols_[:1]).mean().reset_index()[cols_]
temp_rural.rename(columns={"prop_norm": "2020_rural"}, inplace=True)
diff_df_ru = diff_df_ru.merge(temp_rural, on=['week'])
diff_df_ru['change_rural'] = (diff_df_ru["2020_rural"].values - diff_df_ru["2019"].values)/diff_df_ru["2020"].values

temp_urban = devices_per_day[(devices_per_day.year==2020) & (devices_per_day.NRUC11=='Urban')].groupby(cols_[:1]).mean().reset_index()[cols_]
temp_urban.rename(columns={"prop_norm": "2020_urban"}, inplace=True)
diff_df_ru = diff_df_ru.merge(temp_urban, on=['week'])
diff_df_ru['change_urban'] = (diff_df_ru["2020_urban"].values - diff_df_ru["2019"].values)/diff_df_ru["2020"].values

diff_df_ru = diff_df_ru.round(4)
diff_df_ru.head()
```

```python
sns.set(font_scale=2)
sns.set_style('white')
colors_ru = ['#00a3f1','#ff5d8f', '#f5d200']

f, ax = plt.subplots(1, figsize=(16, 5)) 
sns.lineplot(data=diff_df_ru, x="week", y='change_rural', linewidth=3, color=colors_ru[0])
sns.lineplot(data=diff_df_ru, x="week", y='change_urban', linewidth=3, color=colors_ru[2])

plt.axvline(x=13,linestyle='dashed',lw=3,color='#595959',label='1st Lockdown')
plt.xlabel("Week of the year")
plt.ylabel("Baseline (%)")
plt.show()
```

```python
interval = 4
file_name = "figures/duration_trips_ru.svg"
plot_lollipop_ru(diff_df_ru, interval, file_name)
```

```python
green_areas_df = pd.read_csv('files/full_gb_green_visits_norm.csv')
green_areas_df['dt'] = green_areas_df.apply(lambda x: datetime.date(x.year, x.month, x.day), axis=1)
green_areas_df.head()
```

```python
green_areas_df['dt'] = green_areas_df.apply(lambda x: datetime.datetime(x.year, x.month, x.day, x.hour), axis=1)
green_areas_df_19 = green_areas_df[green_areas_df.year==2019]
green_areas_df_19['points_norm'] = (green_areas_df_19['points'] - green_areas_df_19['points'].min())/(green_areas_df_19['points'].max() - green_areas_df_19['points'].min())
green_areas_df_20 = green_areas_df[green_areas_df.year==2020]
green_areas_df_20['points_norm'] = (green_areas_df_20['points'] - green_areas_df_20['points'].min())/(green_areas_df_20['points'].max() - green_areas_df_20['points'].min())
```

```python
cols = ['year', 'week', 'geo_code']
users_count = pd.read_csv("users_count.csv")
green_areas_df = pd.concat([green_areas_df_19, green_areas_df_20])
green_areas_df = green_areas_df.groupby(cols).sum().reset_index()
```

```python
green_areas_week = green_areas_df.groupby(['geo_code', 'year','week']).mean().reset_index()
df_diff_gc = green_areas_week[green_areas_week.year==2019]
df_20 = green_areas_week[(green_areas_week.year==2020) & (green_areas_week.week<53)]   
df_diff_gc['points_diff'] = (df_diff_gc.points_norm.values - df_20.points_norm.values)/df_diff_gc.points_norm.values
df_diff_gc["points_perc"] = df_diff_gc["points_diff"].apply(aux_func)
```

```python
colors_ru = ['#00a3f1','#595959', '#f5d200']
urban_rural_df = pd.read_csv('files/RUC11_LAD11_ENG.csv')
df_diff_gc  = df_diff_gc.merge(urban_rural_df[['geo_code', 'RUC11CD']], on='geo_code')
df_diff_gc.RUC11CD = df_diff_gc.RUC11CD.astype(int)
df_diff_gc['NRUC11'] = df_diff_gc.RUC11CD.apply(lambda x: get_new_ru_text(x))
```

```python
sns.set(font_scale=2)
sns.set_style('white')
f, ax = plt.subplots(1, figsize=(16, 5))
sns.lineplot(data=df_diff_gc, x="week", y="points_perc", hue='NRUC11', linewidth=3, palette=colors_ru)
plt.axvline(x=13,linestyle='dashed',lw=3,color='black',label='1st Lockdown')
plt.xlabel("Week of the year")
plt.ylabel("Baseline (%)")
sns.move_legend(ax, "upper right", ncol=3, title=None, frameon=False,)
plt.savefig("figures/visits_ru_baseline.svg", transparent=True)
plt.show()
```
